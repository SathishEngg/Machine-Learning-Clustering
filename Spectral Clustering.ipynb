```python
from sklearn.cluster import SpectralClustering
from sklearn.metrics import pairwise_distances
X = np.random.rand(100, 2)
num_clusters_list = []
n_clusters_range = np.arange(2, 11)  # Spectral Clustering requires n_clusters > 1
for n_clusters in n_clusters_range:
    spectral_clustering = SpectralClustering(n_clusters=n_clusters, affinity='nearest_neighbors')
    spectral_clustering.fit(X)
    num_clusters = len(np.unique(spectral_clustering.labels_))
    num_clusters_list.append(num_clusters)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(n_clusters_range, num_clusters_list, marker='o', linestyle='--', color='b')
plt.title('Effect of Number of Clusters on Spectral Clustering')
plt.xlabel('Number of Clusters')
plt.ylabel('Number of Clusters Found')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import SpectralClustering
from sklearn.datasets import make_blobs
X, _ = make_blobs(n_samples=200, centers=4, cluster_std=0.60, random_state=42)
n_clusters = 4
spectral_clustering = SpectralClustering(n_clusters=n_clusters, affinity='nearest_neighbors', random_state=42)
y_spectral = spectral_clustering.fit_predict(X)
```

```python
import seaborn as sns
np.random.seed(42)
supervised = pd.DataFrame({
    'CustomerID': np.arange(1, 101),
    'Age': np.random.randint(18, 70, 100),
    'Annual Income (k$)': np.random.randint(15, 140, 100),
    'Spending Score (1-100)': np.random.randint(1, 100, 100),
})
x_col = 'Annual Income (k$)'
y_col = 'Spending Score (1-100)'
spectral_clustering = SpectralClustering(n_clusters=4, affinity='nearest_neighbors', random_state=42)
supervised['Cluster'] = spectral_clustering.fit_predict(supervised[[x_col, y_col]])
```

```python
facet = sns.lmplot(
    data=supervised, 
    x=x_col, 
    y=y_col, 
    hue='Cluster', 
    fit_reg=False, 
    legend=True, 
    legend_out=True, 
    palette="viridis"
)
plt.title('Spectral Clustering')
plt.xlabel(x_col)
plt.ylabel(y_col)
plt.show()
```
